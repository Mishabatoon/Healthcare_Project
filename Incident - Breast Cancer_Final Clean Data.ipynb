```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Import file
file = "Data/Incident_data.txt"
df = pd.read_csv(file, encoding="ISO-8859-1", sep='\t')
display(df)
```

```python
# Drop all columns without any information
df_dropped = df.drop(columns=['Notes','Leading Cancer Sites Code', 'Leading Cancer Sites', 'Race Code', 'States Code', 'Year Code' ],axis=1)
df_dropped = df_dropped.dropna(how='any')
display(df_dropped)
```

```python
df_no_other_races = df_dropped[df_dropped["Race"]!="Other Races and Unknown combined"]
df_no_other_races
```

```python
df_clean_incident = df_no_other_races[df_no_other_races["Year"]!=1999.0]
df_clean_incident
```

```python
df_clean_incident.info()
```

```python
df_clean_incident['Population'] = df_clean_incident['Population'].astype(float)
```
