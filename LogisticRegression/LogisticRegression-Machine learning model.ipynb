## cleaning data

```python
# Dropping columns 
incidence_df.drop(columns= ['Notes','Leading Cancer Sites Code','States Code','Race Code'],inplace=True)
incidence_df
```

```python
# retrieve the array of data
data = incidence_df.values
```

## logistic regression for incidence data

```python
# separate into input and output columns
X = data[:, :-1].astype(str)
y = data[:, 1]
```

```python
# summarize
print('Input', X.shape)
print('Output', y.shape)
```

```python
# split the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.15, random_state=42)
```

```python
# ordinal encode input variables
ordinal = OrdinalEncoder()
X = ordinal.fit_transform(X)
```

```python
# ordinal encode input variables
ordinal_encoder = OrdinalEncoder()
ordinal_encoder.fit(X_train)
X_train = ordinal_encoder.transform(X_train)
X_test = ordinal_encoder.transform(X_test)
```

```python
# define the model
model = LogisticRegression()
# fit on the training set
model.fit(X_train, y_train)
```

```python
# predict on test set
yhat = model.predict(X_test)
```

## Accuracy score result
