```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Import file
file = "Data/Mortality_Rate.txt"
df = pd.read_csv(file, encoding="ISO-8859-1", sep='\t')
display(df)
```

```python
# Drop all columns without any information
df_dropped = df.drop(columns=['Notes','Leading Cancer Sites Code', 'Leading Cancer Sites', 'Race Code', 'State Code', 'Year Code' ],axis=1)
df_dropped = df_dropped.dropna(how='any')
display(df_dropped)
```

```python
df_clean_mortality = df_dropped[df_dropped["Race"]!="Other Races and Unknown combined"]
df_clean_mortality
```
